Utilizando las librerías obtenga la utilización de tarjetas de crédito: por sexo (cuenta) (20%), total en dólares utilizados diferenciando tarjetas de crédito por ciudad (2 criterios) (20%), total utilizado por tipo de gasto (20%). Genere gráficos (usted elija) de los 3 resultados usando matplotlib (20%) Genere como salida 3 archivos Excel, uno por cada resultado (20%)

```python
#por sexo.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Modificamos el archivo por si a usted le da un errror y fueron:
#1-campo de CardType
#2-campo de ExpType
tarjetas = pd.read_csv("/content/trx_tarjetas_credito (1).csv",sep=";",encoding="utf-8")
```

```python
tarjetasPsexo = pd.crosstab(index=tarjetas["Gender"],columns="utilizacion de tarjetas por sexo")
plt.pie(tarjetasPsexo,labels=tarjetasPsexo.index,autopct="%0.lf%%")
plt.xlabel("Total Utilizado por Tipo de Gasto")
plt.show()
tarjetasPsexo.to_csv("/content/tarjetasPsexo.csv",sep=";")
```

```python
#TIPO E TARJETA POR CIUDAD
tarjetasPCiudad = pd.crosstab(index=[tarjetas.City.head(23),tarjetas.CardType],columns="a")
plt.pie(tarjetasPCiudad,labels=tarjetasPCiudad.index,autopct="%0.1f%%")
plt.xlabel("topo de tarjeta por ciudad")
plt.show()
tarjetasPsexo.to_csv("/content/tarjetasPCiudad.csv",sep=";")
```

```python
#TOTAL UTILIZADO POR TIPO DE GASTO
tipoGasto = pd.crosstab(index=tarjetas.ExpType,columns="tipogasto")
plt.pie(tipoGasto,labels=tipoGasto.index,autopct="%0.lf%%")
plt.xlabel("Total Utilizado por Tipo de Gasto")
plt.show()
tipoGasto.to_csv("/content/tipoGasto.csv",sep=";")
```
